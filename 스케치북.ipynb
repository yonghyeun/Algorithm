```python
class Stack:

  def __init__(self):

    self.arr = []

  def push(self,value):
    self.arr.append(value)
  
  def pop(self):
    if self.arr:
      return self.arr.pop()
    else:
      return -1 
  
  def __len__(self):
    return len(self.arr)

  def empty(self):
    if len(self) > 0:
      return 0
    else:
      return 1 
  
  def show(self):
    if self.empty():
      return -1
    else:
      return self.arr[-1]
    
N = int(input())

stack =  Stack()
for _ in range(N):
  
  x = list(map(int,input().split(' ')))

  if len(x) == 2:
    x , value = x[0], x[1]
  else:
    x = x[0]

  if x == 1:
    stack.push(value)
  if x == 2:
    print(stack.pop())
  if x == 3:
    print(len(stack))
  if x == 4:
    print(stack.empty())
  if x == 5:
    print(stack.show())
```

```python
def check(arr : list[str], stack : list[str]):

  while stack and arr: # stack, arr 에 담긴 글자가 모두 사라질 때 까지 반복 
    if stack[-1] == arr[-1]: # stack 에 담긴 글자와 arr의 글자가 같다면 좋은 글자
      stack.pop()
      arr.pop()
    else: # 만약 다르다면
      stack.append(arr.pop())
    
    '''
    반복문이 끝나면 두 가지 경우가 남는다
    1. stack ,arr 모두 비었을 경우 : good string 
    2. arr는 모두 비고 stack에는 값이 남았을 경우 : not good string
    '''

  if stack:
    return 0 
  else:
    return 1


N = int(input())
cnt = 0
for _ in range(N):
  
  string = list(input().rstrip())
  s = [string.pop()] # 초기 스택은 한 값을 일단 pop 해둔 형태
  cnt += check(string , s)
print(cnt)
```

```python
'''
MaxHeap : 부모 노드가 자식 노드보다 값이 항상 크도록 만족하는 자료구조 
heapify_up , heapify_down: 힙 성질을 만족 할 수 있도록 자식 노드가 부모노드보다 값이 크다면
            부모노드와 자식 노드의 값을 변경하는 함수
'''


from collections import deque

def heapify_up(arr : deque[int]) -> deque:
  '''
  값이 insert 되었을 때 leaf node 로 insert 되는데
  insert 된 값의 부모노드들과 값을 비교하며 위치를 바꿔 나가는 함수
  '''

  n = len(arr) # 원소의 개수
  k = n - 1 # 추가된 값의 인덱스 
  while k > 0 and arr[k] > arr[(k-1) // 2]: 
    # root node에 도달하지 않았으면서 부모노드보다 자식 노드의 값이 클 경우 수행 
    arr[k] , arr[(k-1) // 2] = arr[(k-1) // 2] , arr[k]
    k = (k-1 // 2)

  return arr

def heapify_down(arr : deque[int]) -> deque:
  '''
  값이 delete 되었을 때 leaf node의 값을 root node로 옮겨진다.
  옮겨진 root node의 값을 적절한 위치로 집어 넣는 함수 
  '''
  n = len(arr)
  k = n - 1 # 추가된 값의 인덱스
  m = k # 초기 설정 값 

  while k*2 + 1 < n:
    L , R = k*2 + 1 , k*2 + 2
    
    if arr[L] > arr[k]: # k 의 왼쪽 자식 노드가 K 보다 크다면
      m = k 
    if R < n and arr[R] > arr[m]: # 오른쪽 자식 노드가 존재 하면서 오른쪽 자식 노드가 m 보다 크다면
      m = R 

    if m != k: # 만약 값을 바꿔야 한다면
      arr[m] , arr[k] = arr[k] , arr[m] # swap
      k = m # 바뀐 위치에서 값을 재탐색 할 수 있도록 
    else:
      break
      
  return arr


N = int(input())
heap =  deque()

for _ in range(N):
  c = int(input())

  if c == 0:

    if len(heap) == 0:
      print(0)
    else:
      if len(heap) <= 2:
        print(heap.popleft())
      else:
        print(heap.popleft())
        heap.appendleft(heap.pop()) # root node 에 leaf node 값을 넣고 
        heap = heapify_down(heap)
    

  else:
    heap.append(c)
    heap = heapify_up(heap)
```

```python
from collections import deque
N = int(input())

info = {}

for _ in range(N):
  L , H = map(int, input().split())
  info[L] = H

lh_arr = deque(sorted(info.items() , key = lambda x: x[0])) # 인덱스와 기둥의 높이 순 정렬

# 면적을 계산해보자

idx , h = lh_arr.popleft()

area = 0
while idx < lh_arr[-1][0]: # 마지막 지붕 만나기 전까지 면적 계산
  while idx < lh_arr[0][0]: # 다음 기둥 만나기 전까지 면적 계산
    area += h 
  # 반복문이 끝났으면 높이를 결정해야 됨
  next_p , next_h = lh_arr.popleft() # 다음 벽돌 인덱스 및 다음 벽돌 인덱스의 높이  
  if h < next_h: # 다음 벽돌이 현재 높이보다 높으면  
    h = next_h # 면적 계산 할 때 사용할 높이를 변경 

# 반복문이 끝났으면 마지막 기둥 인덱스 이전까지 지붕이 계산 되었음
# 이 때 next_p 는 마지막 기둥 전 제일 높았던 기둥이며, 지붕은 높게 지어졌을 수 있음

last_p , last_h = lh_arr[0]
if h > last_h: # 만약 마지막 기준 기둥이 마지막 기둥보다 높다면
  # 현재 지붕은 과하게 많이 올라간 상태임 
  area -= (h - last_h) * (last_p - next_p -1)

# 끝났으면 마지막 지붕의 끝자락 추가하기

area += last_h 

print(area)
```

```python
N = int(input())
dasom = int(input())

arr = [int(input()) for _ in range(N -1)] 

cnt = 0 

if len(arr) == 0:
  print(0)
else:  
  while True:
    max_value = max(arr)
    max_idx = arr.index(max_value)
    max_person = arr.pop(max_idx)

    if dasom <= max_person:
      dasom += 1
      max_person -= 1
      cnt += 1
      arr.append(max_person)
    else:
      break

  print(cnt)
```
